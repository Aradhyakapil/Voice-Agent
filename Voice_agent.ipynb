```python
import openai
import whisper
from elevenlabs.client import ElevenLabs
from elevenlabs import play
from google.colab import files
from IPython.display import Audio, display

# Replace with your own keys
openai.api_key = "sk-"
tts = ElevenLabs(api_key="")

# Load Whisper model (this may take a minute)
model = whisper.load_model("base")

# â”€â”€â”€ 2. Loop: upload â†’ transcribe â†’ chat â†’ speak â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
while True:
    print("ðŸ”½ Please upload a WAV file (or name it â€˜exit.wavâ€™ to stop):")
    uploaded = files.upload()                   # opens file picker
    fname = next(iter(uploaded.keys()))

    # Exit condition
    if fname.lower().startswith("exit"):
        print("ðŸ‘‹ Exiting.")
        break

    # 2.1 Transcribe with Whisper
    result = model.transcribe(fname)
    prompt = result["text"].strip()
    print(f"ðŸ§‘ You said: {prompt}")

    # 2.2 Query GPT-4o-mini
    resp = openai.chat.completions.create(
        model="gpt-4o",
        messages=[{"role":"user","content":prompt}]
    )
    reply = resp.choices[0].message.content.strip()
    print(f"ðŸ¤– GPT-4o-mini: {reply}")

    # 2.3 Convert reply to speech & play
    audio = tts.text_to_speech.convert(
        text=reply,
        voice_id="21m00Tcm4TlvDq8ikWAM",              # or specific voice_id from voices list
        model_id="eleven_monolingual_v1",
        output_format="mp3_44100_32",   # choose desired output format
    )
    audio_bytes = b"".join(audio)

    # ðŸ”Š Play in Colab
    display(Audio(audio_bytes, autoplay=True))
```
